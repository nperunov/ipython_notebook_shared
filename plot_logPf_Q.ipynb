```python
def get_coordinates(traj):
    x = traj["id"]/Ny
    y = traj["id"]%Ny
    xdiff = np.diff(x)
    xdiff[xdiff==-(Nx-1)] = 1
    xdiff[xdiff==(Nx-1)] = -1

    ydiff = np.diff(y)
    ydiff[ydiff==-(Ny-1)] = 1
    ydiff[ydiff== (Ny-1)] = -1

    x = np.concatenate(([x[0]], x[0]+xdiff.cumsum()))
    y = np.concatenate(([y[0]], y[0]+ydiff.cumsum()))
    return x, y
```

# Finding points of fixed reversal probability

In this section we are going to check if the following relation holds:
\begin{equation}
    \ln \frac{\pi_\tau(i\rightarrow j)}{\pi_\tau(i\rightarrow k)} = 
    \ln \frac{\pi_\tau(j\rightarrow i)}{\pi_\tau(k\rightarrow i)} -
    \ln \frac{\langle e^{-\beta q_{ij}}\rangle_\tau}{\langle e^{-\beta q_{ik}}\rangle_\tau}.
\end{equation}
There are two ways to estimate the l.h.s. of the equation above.
First, one can estimate it by counting the total number of trajectories starting in $i$ and ending in $j$ or $k$, and then taking their ratio:
\begin{equation}
    \ln \frac{\pi_\tau(i\rightarrow j)}{ \pi_\tau(i\rightarrow k)} =
    \ln \frac{N_\tau(i\rightarrow j)}{ N_\tau(i\rightarrow k)}
\end{equation}
Second, one can take the sum of probabilities of indivual trajectories:
\begin{equation}
    \pi_\tau(i\rightarrow j) = \sum_{\text{trajectories}} p(trajectory)
    \approx \max_{\text{trajectories}} p(trajectory)
\end{equation}

# Estimation of exponential averages for Gaussian distibution

Lets assume that we know a multivariable Gaussian distribution for vector ${\bf x}$
\begin{equation}
    PDF({\bf x} ) = \sqrt{\frac{det~K} {\pi^N}  }  \cdot \exp[ -{\bf x}^{T} K {\bf x}],
\end{equation}
where $K$ is positively defined matrix of size $N.$ Then the average of $e^{{\bf k}{\bf x}}$ is given by 
\begin{equation}
    \langle e^{{\bf k}{\bf x}} \rangle = \exp[ \frac{1}{4}{\bf k}^{T} K^{-1} {\bf k} ].
\end{equation}
For two dimensional case ${\bf x} = (x_1, x_2),~{\bf k} = (k_1, k_2),$ and $K = \pmatrix{a & c\\ c & b}$, one can find that
\begin{align}
    &K^{-1} = \frac{1}{ab - c^2}\pmatrix{b & -c\\ -c & a} \\
    &\sigma_{x_1} = \frac{1}{2}\frac{b}{ab - c^2} \\
    &\sigma_{x_2} = \frac{1}{2}\frac{a}{ab - c^2} \\
    &\mathrm{cov}(x_1, x_2) = \frac{1}{2}\frac{-c}{ab - c^2}\\
    &\rho = \mathrm{corr}(x_1, x_2) = \frac{-c}{\sqrt{ab}} \\
    &\langle e^{{\bf k}{\bf x}} \rangle = \exp\big{[} \frac{1}{4(ab - c^2)} (  b k_1^2 + a k_2^2 - 2c k_1 k_2 )  \big{]} = 
    \exp\big{[}  (  \sigma_{x_1}^2 k_1^2 + \sigma_{x_2}^2 k_2^2 + 2 \mathrm{cov}(x_1, x_2) k_1 k_2 )/2  \big{]}.
\end{align}

We are interested in computing $\langle e^{-Q} \rangle_{ij},$ assuming that $\ln P_f$ and $Q$ is a pair of correlated Gaussian variables.
\begin{align}
    \langle e^{-Q} \rangle_{ij} &= \frac{\sum_{x(t)} P_f[x(t)] e^{-Q[x(t)]}}{\sum_{x(t)} P_f[x(t)]  } = \\
                                &= \frac{\sum_{x(t)} e^{\ln P_f[x(t)] - Q[x(t)]} } { \sum_{x(t)} e^{\ln P_f[x(t)]} } =\\
             &= \frac{ e^{-\overline{Q}}\langle e^{(1,-1)(\Delta\ln P_f, \Delta Q)} \rangle } {\langle e^{(1,0)(\Delta\ln P_f, \Delta Q)} \rangle} =\\
             & = \frac{  e^{-\overline{Q}}  \exp\big{[}  (  \sigma_{\ln P_f}^2 + \sigma_{Q}^2 - 2 \mathrm{cov}(\ln P_f, Q) )/2  \big{]}  }{ \exp\big{[}  \sigma_{\ln P_f}^2  /2  \big{]}} = e^{-\overline{Q}} \exp \big{[}   \sigma_{Q}^2/2 - \mathrm{cov}(\ln P_f, Q)   \big{]}
\end{align}
