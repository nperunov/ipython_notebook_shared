```python
trajectory_dtype=[('id', int),('artime', int),('deptime', int),
       ('dW', np.float64), ('dlogPf', np.float64),
       ('dlogPr', np.float64), ('dQ', np.float64)]

def load_trajectory( filename ):
    traj = np.genfromtxt( filename, trajectory_dtype )
    return traj

def compute_pos_W_logPf_logPr_Q( trajectory ):
    njumps = trajectory.shape[0] if trajectory.shape else 1
    pos = (trajectory['id'][-1] if njumps > 1 else float(trajectory['id']))
    W = trajectory['dW'].sum()
    logPf = trajectory['dlogPf'].sum()
    logPr = trajectory['dlogPr'].sum()
    Q = trajectory['dQ'].sum()
    return pos, W, logPf, logPr, Q, njumps

def fill_missing_values(count):
    count_tmp = pd.Series(np.zeros(1600, dtype=float), name='pos')
    count_tmp.ix[count.index] = count
    return count_tmp
    
```

```python
def get_coordinates(traj):
    x = traj["id"]/Ny
    y = traj["id"]%Ny
    xdiff = np.diff(x)
    xdiff[xdiff==-(Nx-1)] = 1
    xdiff[xdiff==(Nx-1)] = -1

    ydiff = np.diff(y)
    ydiff[ydiff==-(Ny-1)] = 1
    ydiff[ydiff== (Ny-1)] = -1

    x = np.concatenate(([x[0]], x[0]+xdiff.cumsum()))
    y = np.concatenate(([y[0]], y[0]+ydiff.cumsum()))
    return x, y
```

# Finding points of fixed reversal probability

In this section we are going to check if the following relation holds:
\begin{equation}
    \ln \frac{\pi_\tau(i\rightarrow j)}{\pi_\tau(i\rightarrow k)} = 
    \ln \frac{\pi_\tau(j\rightarrow i)}{\pi_\tau(k\rightarrow i)} -
    \ln \frac{\langle e^{-\beta q_{ij}}\rangle_\tau}{\langle e^{-\beta q_{ik}}\rangle_\tau}.
\end{equation}
There are two ways to estimate the l.h.s. of the equation above.
First, one can estimate it by counting the total number of trajectories starting in $i$ and ending in $j$ or $k$, and then taking their ratio:
\begin{equation}
    \ln \frac{\pi_\tau(i\rightarrow j)}{ \pi_\tau(i\rightarrow k)} =
    \ln \frac{N_\tau(i\rightarrow j)}{ N_\tau(i\rightarrow k)}
\end{equation}
Second, one can take the sum of probabilities of indivual trajectories:
\begin{equation}
    \pi_\tau(i\rightarrow j) = \sum_{\text{trajectories}} p(trajectory)
    \approx \max_{\text{trajectories}} p(trajectory)
\end{equation}


```python
pi_forward = grouped.apply(lambda x: x.logPf - x.logPf.mean() )
pi_reverse = grouped.apply(lambda x: x.logPr - x.logPr.mean() )

pi_f_max = grouped.apply(lambda x: x.logPf.max() )
pi_r_max = grouped.apply(lambda x: x.logPr.max() )
Q_min = grouped.apply(lambda x: x.Q.min() )
expQ = grouped.apply(lambda x: np.log((np.exp(x.logPf-x.Q)).sum()/np.exp(x.logPf).sum() ))

#print pi_f_max[0], pi_r_max[0]

plt.figure(figsize=(10,10))
plt.subplot(221)
plt.imshow(fill_missing_values(pi_f_max).reshape(Nx, Ny).T, origin='lower', interpolation='Nearest')
plt.colorbar()
plt.subplot(222)
plt.imshow(fill_missing_values(pi_r_max).reshape(Nx, Ny).T, origin='lower', interpolation='Nearest')
plt.colorbar()
plt.subplot(223)
plt.imshow(fill_missing_values(expQ).reshape(Nx, Ny).T, origin='lower', interpolation='Nearest')
plt.colorbar()
plt.subplot(224)
plt.imshow(fill_missing_values((pi_f_max-pi_r_max)).reshape(Nx, Ny).T, origin='lower', interpolation='Nearest')
plt.colorbar()
```
